```python
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

# Create a Wordcloud Image

```python
# Create a list of words

text=(
    'Python Python Python Matplotlib Matplotlib Seaborn Network Plot Violin \
    Chart Pandas Datascience Wordcloud Spider Radar Parrallel Alpha Color Brewer \
    Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart \
    Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble'
)
```

```python
text
```

```python
# Create the wordcloud object

wordcloud = WordCloud(width=480, height=480, margin=0).generate(text)
```

```python
# Display the generated image:

plt.imshow(wordcloud, interpolation='bicubic')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

# Save plot as Image

```python
# To save plt as image please write code 'plt.savefig('wordcloudimage1.jpg')' before 'plt.show()'

plt.imshow(wordcloud, interpolation='bicubic')
plt.axis("off")
plt.margins(x=0, y=0)
plt.savefig('wordcloudimage1.jpg')
plt.show()
```
